```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Basics of matplot lib

```python
x = np.linspace(0,5,11)
x
```

```python
y = x ** 3
y
```

```python
plt.plot(x,y)
plt.title("Basic Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")
```

#Sub plots in Matplotlib

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(y**2,y**2)
```

# Object Orented Way

```python
# Create an empty figure (canvas)
fig = plt.figure(figsize=(20,10),dpi=250,facecolor='lightblue')

# Add first axes at [left, bottom, width, height]
ax1 = fig.add_axes([0.5,0.1,0.4,0.4])
ax1.plot(x,y)

# Add second axes at a different position
ax2 = fig.add_axes([0.5,0.5,0.3,0.3])
ax2.plot(x,y)

# Display the updated figure
plt.show()
```

# @title
# [left, bottom, width, height
# position of axes + size of axes (0 to 1 relative to figure)

Types of Plot

Saving Plots

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("baseplot.png")
```
