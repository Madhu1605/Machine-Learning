```python
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

• distplot/histplot
• jointplot
• pairplot
• rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.displot(df['total_bill'],bins = 20,kde = True) #displot and histplot are same kde = kernal density estimeter
```

```python
plt.subplot(1,2,1)
sns.distplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.distplot(df['tip'],bins = 20,kde = True)
```

Joint Plot is mainly used to compare the relationship between two numerical variables. 👍

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind = 'scatter')
```

Pairplot is used to compare all numerical columns with each other at once.

Suppose your dataset has:
Age
Salary
Experience
Marks

Pairplot automatically creates:

Age vs Salary
Age vs Experience
Age vs Marks

Salary vs Experience
Salary vs Marks

Experience vs Marks

And also shows the distribution of each column on the diagonal.

```python
sns.pairplot(df,hue = 'sex')
```

```python
sns.pairplot(df,hue = 'size' , palette = 'rainbow')
```

Rug Plot

A rug plot shows every data point as a small line (tick mark) on an axis.

```python
sns.rugplot(df)
```

Categorical Data Plots

Now let's discuss using seaborn to plot categorical data! There are a few main plot types for this:

- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

Let's go through examples of each!

Count Plot

A countplot shows the count (frequency) of each category.

when to use :- Use a count plot when your question starts with:"How many...?"

```python
sns.countplot( x = df['sex'],hue = df['smoker'])
```

Bar Plot

A bar plot compares a numerical value across categories.

Use a bar plot when your question starts with:

What is the average salary?
Which category has the highest mean marks?

```python
import numpy as np
sns.barplot(x = df['sex'] , y=df['tip'],estimator = np.sum)
```

```python

sns.barplot(x = df['sex'] , y=df['total_bill'],estimator = np.sum)
```

 BOX PLOT

```python
sns.boxplot(x = df['tip'] , y = df['day'] , data = df, palette = 'rainbow')
```

boxplot and violinplot

boxplots and violinplots are used to show the distribution of categorical data.

A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.

The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be "outliers" using a method that is a function of the inter-quartile range.

```python
sns.violinplot(y = df['total_bill'],x = df['day'] ,data = df)
```

stripplot and swarmplot

The stripplot will draw a scatterplot where one variable is categorical.
A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don't overlap.
This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations.

```python
sns.stripplot(x = df['day'] , y = df['total_bill'] , data = df)
```

```python
sns.swarmplot(x = df['day'] , y = df['total_bill'] , data = df)
```

```python
sns.violinplot(x = df['day'] , y = df['total_bill'],data = df)
sns.swarmplot(x = df['day'] , y = df['total_bill'],data = df)
```
