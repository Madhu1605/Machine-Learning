```python
import numpy as np
import pandas as pd

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}

df = pd.DataFrame(data)

print(df)
```

```python
# Checks every value:

# True → value is missing (NaN)
# False → value exists

df.isna()
```

```python
# Counts total missing values column-wise.

df.isna().sum()
```

```python
# Checks whether a column contains at least one NaN.
df.isna().any()
```

```python
# REMOVING MISSING DATA
```

```python
df.dropna() #Removes rows that contain at least one NaN
            #Every row has a missing value
            #Empty DataFrame
```

```python
df
```

```python
df.dropna(thresh=1)
df
#thresh means:
#minimum number of NON-NaN values required to keep the row.
#At least 1 non-NaN value must exist.
#All rows satisfy this ✔️
```

```python
# FILLING THE MISSING DATA
# remove -> dropna() row wise
# fill -> fillna() col wise
#search -> isna() col wise

df
```

```python
df.fillna(0,inplace=True)
df
```

```python
df
```

```python
val = {"A":100,"B":200,"C":300,"D":400}
df.fillna(value=val,inplace=True)
df
# Column A → Fill NaN with 100
# Column B → Fill NaN with 200
# Column C → Fill NaN with 300
# Column D → Fill NaN with 400
```

```python
# FIlling the values with mean
df.fillna(df.mean(),inplace=True)
df
```

```python
import pandas as pd

# DataFrame 1: Employee information
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4, 5],
    'name': ['John', 'Anna', 'Peter', 'Linda', 'Bob'],
    'department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

# DataFrame 2: Salary information
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'salary': [60000, 80000, 65000, 70000, 90000],
    'bonus': [5000, 10000, 7000, 8000, 12000]
})
```

```python
pd.merge(employees,salaries,on="employee_id") #(default is inner)
```

```python
pd.merge(employees,salaries,on="employee_id",how='inner')
```

```python
pd.merge(employees,salaries,on="employee_id",how='outer') #Keeps ALL rows from BOTH DataFrames.
```

```python
pd.merge(employees,salaries,on="employee_id",how='left') #Keeps ALL rows from LEFT DataFrame (employees).
                                                        #If no match exists → NaN.
```

```python
pd.merge(employees,salaries,on="employee_id",how='right') #Keeps ALL rows from RIGHT DataFrame (salaries).
```

```python
import pandas as pd

data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}

df = pd.DataFrame(data)

print(df)
```

```python
# Group by Category and Calculate the Sum of Sales
cat = df.groupby("Category")['Sales'].sum()
cat
```

```python
# Group by store and Calculate the Sum of the Sales
ca = df.groupby("Store")['Sales'].sum()
ca
```

```python
#Group by Multiple Columns
#Group by Category and Store
cat = df.groupby(['Category','Store'])['Sales'].sum()
cat
```

```python
#Aggrigation
```

```python
df['Sales'].mean()
print(df['Sales'].mean())
#Mean,Median,Mode,Min,Max,count,Std
```

```python
h=df['Sales'].agg(['mean','max','min','sum','median','count','std'])
print(h)
```

```python
#Pivot Tables
A pivot table is used to summarize and reorganize data. It groups the data and calculates statistics like sum, mean, count, max, min, etc.
```

```python
import pandas as pd
import numpy as np

data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': ['East', 'West', 'North', 'South',
               'East', 'West', 'North', 'South',
               'East', 'West', 'North', 'South',
               'East', 'West', 'North', 'South',
               'East', 'West', 'North', 'South'],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': ['John', 'Mary', 'Bob', 'Alice',
            'John', 'Mary', 'Bob', 'Alice',
            'John', 'Mary', 'Bob', 'Alice',
            'John', 'Mary', 'Bob', 'Alice',
            'John', 'Mary', 'Bob', 'Alice']
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)

print(df)
```

```python
pivot1 = pd.pivot_table(df,values='Sales',index='Region')
pivot1
```

```python
pivot2 = pd.pivot_table(df,values='Sales',index='Region',columns='Product')#aggfun = meadian mean
pivot2
```

```python
pivot2 = pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')#aggfun = meadian mean
pivot2
```

```python
#Cross Tabs
#crosstab() is used to count the frequency (number of occurrences) between two or more categorical variables.
```

```python
pd.crosstab(df['Region'],df['Product'])
```
