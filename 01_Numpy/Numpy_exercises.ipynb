```python
import numpy as np

# Columns: [Age, Math Marks, Science Marks]
data = np.array([
    [18, 85, 78],   # Student 1
    [19, 92, 88],   # Student 2
    [17, 76, 95],   # Student 3
    [18, 65, 70],   # Student 4
    [20, 90, 85]    # Student 5
])
```

Get the shape of the matrix.

```python
print(data.shape)
```

Find the average age of students.

```python
avg = data[0:,0]
print(avg.mean())
```

Extract Math marks of all students.

```python
maths = data[:,1]
print(maths)
```

Find the highest Science mark.

```python
science = data[:,2]
print(science.max())
```

```python
print(np.max(data[:,2]))
```

Get details of the student who scored more than 90 in Math.

```python
print(data[data[:,1] > 90])
```

Increase Math marks of all students by 5.

```python
print(data[:,1]+5)
```

Find how many students are younger than 19.

```python
age = data[0:,0:1]
count = 0
for i in age:
  if i < 19:
    count = count +1
print(count)
```

```python
print(len(data[data[:,0] < 19]))
```

Calculate the average marks in each subject (column-wise mean).

```python
print(np.mean(data[:,1:],axis=0))
```

Get data of students who scored at least 80 in both subjects.

```python
data[(data[:,1] >= 80) & (data[:,2] >= 80)]
```

Replace all Science marks < 75 with 0.

```python
data[:,2] [data[:,2] < 75] = 0
data
```
